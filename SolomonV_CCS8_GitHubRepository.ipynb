```python
from sklearn.feature_selection import mutual_info_classif as mc
from sklearn.feature_selection import RFE as rf
from sklearn.linear_model import LogisticRegression as lr
from sklearn.preprocessing import MinMaxScaler as mm

import matplotlib.pyplot as pp
import pandas as pd
import seaborn as sb

fi = "https://raw.githubusercontent.com/vjs0303/SolomonV_CCS8_GitHubRepository/main/SolomonV_CCS8_Dataset_Original.csv"
df = pd.read_csv(fi)

sc = mm()

q1 = df['banner_days'].quantile(0.25)
q3 = df['banner_days'].quantile(0.75)
ir = q3 - q1
lb = q1 - 1.5 * ir
ub = q3 + 1.5 * ir

olb = df[(df['banner_days'] < lb) | (df['banner_days'] > ub)]

df['banner_days'] = df['banner_days'].clip(lower=lb, upper=ub)

df = df.drop(columns=['version', 'version_name', 'start_date', 'end_date', 'avg_revenue'])

df['revenue'] = sc.fit_transform(df[['revenue']])
df['banner_days'] = sc.fit_transform(df[['banner_days']])

df = pd.get_dummies(df, columns=['5_star_characters'], drop_first=True)

bo = df.select_dtypes(include=['bool']).columns

df[bo] = df[bo].astype(int)

df.to_csv('SolomonV_CCS8_Dataset_Processed.csv', index=False)
sc = mm()

q1 = df['banner_days'].quantile(0.25)
q3 = df['banner_days'].quantile(0.75)
ir = q3 - q1
lb = q1 - 1.5 * ir
ub = q3 + 1.5 * ir

olb = df[(df['banner_days'] < lb) | (df['banner_days'] > ub)]

df['banner_days'] = df['banner_days'].clip(lower=lb, upper=ub)

df['revenue'] = sc.fit_transform(df[['revenue']])
df['banner_days'] = sc.fit_transform(df[['banner_days']])

if '5_star_characters' in df.columns:
    df = pd.get_dummies(df, columns=['5_star_characters'], drop_first=True)

bo = df.select_dtypes(include=['bool']).columns

df[bo] = df[bo].astype(int)

df.to_csv('SolomonV_CCS8_Dataset_Processed.csv', index=False)
```
